## Task 3: 
Consider a probability distribution of $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$.

Question 1: Find a general formula for the expected value of this distribution.

Question 2: A certain type of tree has seedlings randomly dispersed in a large area, with the mean density of seedlings being approximately five per square yard. If a forester randomly locates ten 1-square-yard sampling regions in the area, find the probability that none of the regions will contain seedlings.

### Question 1:

To find the expected value of this distribution, start with the definition of expected value:

$E[y] = \sum_{y =0}^{\infty} y \cdot p(y)$

Now, substituting the given probability distribution $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$:

$= \sum_{y =0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda}$

We can simplify this expression further:

$= e^{-\lambda} \sum_{y=0}^{\infty} \frac{\lambda^y}{(y-1)!}$

$= e^{-\lambda} \sum_{y=1}^{\infty} \frac{\lambda^y}{(y-1)!}$

$= e^{-\lambda} \sum_{y=1}^{\infty} \frac{\lambda^{y-1}}{(y-1)!}$

Now, notice that the sum is just the Taylor series expansion of $e^{\lambda}$, which is $\sum_{n=0}^{\infty} \frac{\lambda^n}{n!}$.

Using this, we can rewrite the expression for $E[y]$ as:

$E[y] = e^{-\lambda} \cdot (\lambda \cdot e^{\lambda})$

$E[y] = \lambda$

So, the general formula for the expected value of the given distribution is $E[y] = \lambda$.


### Question 2: 

To find the probability that none of the ten 1-square-yard sampling regions will contain seedlings, we can use the Poisson distribution, assuming that the distribution of seedlings follows a Poisson distribution with a mean density of five per square yard.

The Poisson probability mass function is given by:

$P(X = k) = \frac{e^{-\lambda}\lambda^{k}}{k!}$

where:
- $ P(X = k) $ is the probability of observing $ k $ seedlings in a given region.
- $ \lambda $ is the mean number of seedlings per region.
- $ e $ is Euler's number, approximately equal to 2.71828.
- $ k! $ is the factorial of $ k $.

Given that the mean density of seedlings is five per square yard, the mean number of seedlings per 1-square-yard region ($ \lambda $) is also five.

We want to find the probability that none of the regions will contain seedlings, which means $ k = 0 $.

So, plugging the values into the Poisson probability mass function:

$P(X = 0) = \frac{e^{-5} \cdot 5^{0}}{0!}$

$P(X = 0) = e^{-5}$

Now, we can calculate this probability:

$P(X = 0) \approx e^{-5} \approx 0.0067$

Therefore, the probability that none of the ten 1-square-yard sampling regions will contain seedlings is approximately $0.0067$.
