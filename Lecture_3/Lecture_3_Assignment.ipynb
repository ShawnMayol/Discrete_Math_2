$\textbf{Graphs}$
-
- Consists of vertices and edges.

- Various graph components such as loops, parallel edges, adjacent vertices, incident edges, and isolated vertices.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the number of nodes and seed for reproducibility
n = 4  # Number of nodes
seed = 1  # Seed for random number generators for reproducibility

# Generate a random graph with specified number of nodes and edges
G = nx.erdos_renyi_graph(n, 0.3, seed=seed)

# Add self-loop
edgelist = [(0, 0)]
G.add_edges_from(edgelist)

# Generate layout for nodes using spring layout algorithm with a specified seed
pos = nx.spring_layout(G, seed=seed)

# Draw the graph
nx.draw(G, pos=pos, with_labels=True, node_color='skyblue', node_size=1000, font_size=15, font_weight='bold')
plt.show()
```

## $Explanation:$

$\textbf{Vertex Set} = {v_0, v_1, v_2, v_3}$

$\textbf{Edge Set} = {e_0, e_1, e_2}$

$\textbf{Edge-Endpoint Function}$
\begin{gather}
\begin{bmatrix}
\begin{array}{c|c}
Edge & Endpoints \\
e_0 & {v_0} \\
e_1 & {v_0, v_1, v_2} \\
e_2 & {v_0, v_1, v_2} \\
\end{array}
\end{bmatrix}
\end{gather}

- $e_1, e_2$ are adjacent to each other.

- $v_1, v_2$ are adjacent to $v_0$.

- $v_0$ is adjacent to itself.

- $e_1, e_2$ are incident to $v_1$.

- $e_0$ is a loop.

- $v_3$ is an isolated vertex.

$\textbf{Directed Graph (Digraphs)}$
-
- A directed graph (digraph) where edges have a direction.

- Example of a directed graph and explanation of nodes, edges, and their directional relationships.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add nodes
nodes = ['A', 'B', 'C', 'D', 'E']
G.add_nodes_from(nodes)

# Add edges
edges = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E'), ('E', 'A')]
G.add_edges_from(edges)

# Draw the directed graph
plt.figure(figsize=(8, 6))
pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, font_size=15, font_weight='bold', arrowsize=20)
plt.show()
```

## $Explanation:$

In a directed graph, also known as a digraph, edges have a direction associated with them. This means that the relationship between nodes is asymmetric, unlike in undirected graphs where edges represent symmetric relationships.

In the example graph:

- Node A has an outgoing edge directed towards node B.
- Node B has an outgoing edge directed towards node C.
- Node C has an outgoing edge directed towards node D.
- Node D has an outgoing edge directed towards node E.
- Node E has an outgoing edge directed towards node A.

These directed edges represent a one-way relationship from one node to another. For instance, in the edge (A, B), node A is the source or starting node, and node B is the target or ending node. This means that there is a directed path from node A to node B, but not necessarily in the reverse direction.

Directed graphs are used to model various real-world scenarios where relationships have a specific direction or flow, such as transportation networks, social networks with directed connections (followers/following), and flow networks in computer science.

$\textbf{Bipartite Graphs}$
-

- A bipartite graph is a graph whose vertices can be divided into two disjoint sets $ U $ and $ V $ such that every edge connects a vertex in $ U $ to a vertex in $ V $. In other words, there are no edges between vertices in the same set.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a complete bipartite graph
G = nx.complete_bipartite_graph(3, 3)

# Define nodes for each set
nodes_set1 = [1, 2, 3]
nodes_set2 = ['A', 'B', 'C']

# Rename nodes according to sets
mapping = {i: v for i, v in enumerate(nodes_set1)}
mapping.update({i + 3: v for i, v in enumerate(nodes_set2)})
G = nx.relabel_nodes(G, mapping)

# Draw the complete bipartite graph
plt.figure(figsize=(6, 4))
pos = nx.bipartite_layout(G, nodes_set1)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=15, font_weight='bold', arrows=False)
plt.show()
```

## $Explanation:$

In the provided example, we have a bipartite graph consisting of two disjoint sets of vertices:
- Set $ U = \{1, 2, 3\} $ representing vertices from the numeric set.
- Set $ V = \{A, B, C\} $ representing vertices from the alphabetic set.

- **Vertices**: The graph consists of two sets of vertices, where each vertex belongs to either set $ U $ or set $ V $.
  - Set $ U $: Contains vertices 1, 2, and 3.
  - Set $ V $: Contains vertices A, B, and C.

- **Edges**: Each edge connects a vertex from set $ U $ to a vertex from set $ V $. In this bipartite graph, there are edges connecting each vertex from set $ U $ to each vertex from set $ V $, forming a complete bipartite graph.
  - For example, there is an edge between vertex 1 (from set $ U $) and vertex A (from set $ V $), as well as between vertex 2 (from set $ U $) and vertex A (from set $ V $), and so on.

This bipartite graph exemplifies the concept of bipartite graphs, where vertices are divided into two disjoint sets, and edges only connect vertices from different sets.

$\textbf{Degree of a Vertex}$
-

- The degree of a vertex in a graph is the number of edges incident on that vertex

- Loop edges are counted twice. 

- The total degree of a graph is the sum of the degrees of all its vertices. 

```python
import matplotlib.pyplot as plt
import networkx as nx

# Set seed for reproducibility
seed = 96

# Create a graph
G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4), (4, 4)])

# Draw the graph
plt.figure(figsize=(5, 5))
pos = nx.spring_layout(G, seed=seed)
nx.draw(G, pos, with_labels=True, node_size=1000, node_color='lightblue', font_size=15, font_weight='bold', arrows=False)
plt.show()
```

$\textbf{The Handshake Theorem}$
-

The Handshake Theorem, also known as the Handshaking Lemma, states that in any finite undirected graph, the sum of the degrees of all vertices is twice the number of edges. Mathematically, it can be expressed as:

$$
\sum_{v \in V} \text{deg}(v) = 2|E|
$$

where:
- $\sum_{v \in V} \text{deg}(v)$ represents the sum of the degrees of all vertices in the graph $G$,
- $|E|$ denotes the number of edges in the graph $G$.

This theorem is called the Handshake Theorem because it can be visualized as a scenario where each vertex represents a person and each edge represents a handshake between two people. The total number of handshakes is equal to the sum of the degrees of all vertices.


```python
import matplotlib.pyplot as plt
import networkx as nx

# Set seed for reproducibility
seed = 69

# Create a graph
G = nx.Graph()
edges = [('A', 'B'), ('A', 'C'), ('B', 'C'), ('C', 'D'), ('D', 'E')]
G.add_edges_from(edges)

# Draw the graph
plt.figure(figsize=(6, 4))
pos = nx.spring_layout(G, seed=seed)
nx.draw(G, pos, with_labels=True, node_size=1000, node_color='lightblue', font_size=15, font_weight='bold', arrows=False)
plt.show()
```

## $Explanation:$

For the given graph:
- Vertices and their degrees:
    - Vertex A: Degree = 2
    - Vertex B: Degree = 2
    - Vertex C: Degree = 3
    - Vertex D: Degree = 2
    - Vertex E: Degree = 1

    Sum of the degrees of all vertices:
    $
    2 + 2 + 3 + 2 + 1 = 10
    $

- Number of edges in the graph:
    $
    |E| = 5
    $

According to the Handshake Theorem, the sum of the degrees of all vertices should be twice the number of edges. In this case, the sum of the degrees of all vertices is 10, which indeed equals twice the number of edges (5).

    - A path is a trail with no repeated vertices.

    - A circuit is a closed walk with no repeated edges (except for the first and last vertex).

    - A simple circuit is a circuit with no repeated vertices (except for the first and last vertex).

$\textbf{Euler's Trail}$
-

- **Euler's Trail (or Eulerian Path):** A trail in a graph that visits every edge exactly once.

- **Conditions:** For an Eulerian path to exist in a graph, exactly zero or two vertices have odd degree.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Set seed for reproducibility
seed = 42

# Create a graph with vertices having odd degrees
G = nx.Graph()
edges = [(1, 2), (1, 3), (2, 3), (2, 4), (3, 4)]
G.add_edges_from(edges)

# Draw the graph
plt.figure(figsize=(6, 4))
pos = nx.spring_layout(G, seed=seed)
nx.draw(G, pos, with_labels=True, node_size=1000, node_color='lightblue', font_size=15, font_weight='bold', arrows=False)
plt.show()
```

## $Explanation:$

**Vertex 1**: Degree $2$

**Vertex 2**: Degree $3$

**Vertex 3**: Degree $3$

**Vertex 4**: Degree $2$

- The vertices with odd degrees are vertices $2$ and $3$.
- According to Euler's theorem, for an Eulerian trail to exist, the graph must have exactly zero or two vertices with odd degrees.
- In this graph, there are exactly two vertices with odd degrees (vertices $2$ and $3$), which means an Eulerian trail **exists**.

An example of an Eulerian trail in this graph could be:
$3 \rightarrow 4 \rightarrow 2 \rightarrow 3 \rightarrow 1 \rightarrow 2$

$\textbf{The Travelling Salesman Problem}$
-

- It seeks to find the shortest possible route that visits each city exactly once and returns to the original city.
- Cities are represented as vertices, and the distances between them as edges in a complete graph.
- The objective is to minimize the total distance traveled while visiting all cities exactly once.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Use seed for reproducibility
seed = 90

G = nx.Graph()

# Add nodes
G.add_nodes_from(["A","B","C","D", "E", "F"])

# Add self-loops to the remaining nodes
edgelist = [("A","B"),("A","C"),("B","C"),("C","D"),("A","C"),("B","D"),("F","E",),("E","A"),("B","F")]
G.add_edges_from(edgelist)
pos = nx.planar_layout(G)  # Seed for reproducible layout
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={('A', 'B'): '75', 
                 ('A', 'C'): '50', 
                 ('D', 'B'): '40', 
                 ('A', 'E'): '100',
                 ('B', 'C'): '25',
                 ('B', 'F'): '15',
                 ('E', 'F'): '20',
                 ('C', 'D'): '15',},
    font_color='red'
)
nx.draw(G, pos=pos,with_labels=True)
nx.draw(G, pos, width=1, linewidths=1, node_size=1000, node_color='lightblue', alpha=0.9, with_labels=True)

plt.show()
```

## $Explanation:$

**Best Route:** $A \rightarrow C \rightarrow D \rightarrow B \rightarrow F \rightarrow E \rightarrow A$

**Calculation:** $50 + 15 + 40 + 15 + 20 + 100 = 240$

- **Best Route:** The best route represents the optimal path that the traveling salesman should take to visit each city exactly once and return to the starting city.

- **Calculation Process:**

    - Start at City A: Begin the route at City A.
    - Visit City C: Move to City C from City A. This segment adds 50 to the total distance.
    - Move to City D: Proceed to City D from City C. This segment adds 15 to the total distance.
    - Visit City B: Next, go to City B from City D. This segment adds 40 to the total distance.
    - Visit City F: Move to City F from City B. This segment adds 15 to the total distance.
    - Move to City E: Proceed to City E from City F. This segment adds 20 to the total distance.
    - Return to City A: Finally, return to City A from City E, completing the loop. This segment adds 100 to the total distance.
- **Total Distance:** Summing up the distances for each segment yields the total distance traveled by the salesman, which in this case is 240 units.

This calculation method ensures that each city is visited exactly once while minimizing the total distance traveled.
